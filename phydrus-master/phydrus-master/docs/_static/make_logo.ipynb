# Creating the Phydrus Logo with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
from matplotlib import cm
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.view_init(10, 280)

X = np.linspace(0, 1, 250)
Z = np.linspace(0, 1, 250)

X, Z = np.meshgrid(X, Z)
Y = np.zeros_like(X)


x = np.ones((20,20)) * np.linspace(0, 1, 20)
y = np.ones((20,20)) * np.linspace(0, 1, 20)
z = np.ones((20,20)) * np.linspace(0, 1, 20)


bottom = cm.get_cmap('Oranges', 128)
top = cm.get_cmap('Blues_r', 128)

newcolors = np.vstack((top(np.linspace(0, 1, 128)),
                       bottom(np.linspace(0, 1, 128))))
newcmp = ListedColormap(newcolors, name='OrangeBlue')
ax.plot_surface(X, Y, Z, cmap=newcmp, linewidth=0, antialiased=False, zorder=-10)

#draw cube
r = [0, 1]
for s, e in combinations(np.array(list(product(r,r,r))), 2):
    if np.sum(np.abs(s-e)) == r[1]-r[0]:
        if (e[0] == 0) and (e[1] == 1):
            zorder=-10
        elif (e[1] == 1) and (e[1] == 1):
            zorder=-10
            
        else:
            zorder=10
        ax.plot3D(*zip(s,e), color="k", zorder=zorder)
        
ax.text(-0.02, 0.28, 0.29, "PHYDRUS", fontsize=40, fontweight="bold", fontfamily='monospace', color=newcmp(1), zdir="x")
ax.plot([0.03, 0.97], [0.,0], [0.43,0.41], zorder=10, c="w", linewidth=3)
ax.plot([0.03, 0.97], [0.,0], [0.43,0.41], zorder=10, c=newcmp(1), linewidth=1)

# Maybe add the water droplet
# https://math.stackexchange.com/questions/51539/a-math-function-that-draws-water-droplet-shape

ax.grid(False)
ax.set_ylim(0, 1)
ax.set_xlim(0, 1)
ax.set_zlim(0, 1)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.set_axis_off()

plt.savefig("logo.png", dpi=300, bbox_inches="tight")
```
